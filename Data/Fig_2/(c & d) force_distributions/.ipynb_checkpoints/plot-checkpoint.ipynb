```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from cycler import cycler
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
mpl.rcParams['lines.linewidth'] =1.5
mpl.rcParams['markers.fillstyle']='none'
#mpl.rcParams['lines.linestyle'] = '--'
filename='force_distributions'
df=pd.read_csv(filename+'.csv')
dic_diff={"R":['Rigid','#FFA500'],"D4":["Lipid",'#15B01A']}

colors_dic={"RNR":"0.5","RR":"0.5","D4NR":"darkgreen","D4R":"darkgreen"}
marker_size=6
dic_rotation={"R":['With rotation','--','^'],"NR":["Without rotation",'--','o']}
sim_list=["AS16"+key1+"H"+key2+"_v3" for key1 in dic_diff.keys() for key2 in dic_rotation.keys()]

def bootstapper(arr,minf,maxf,no_bins):
    boot_strap_sample_size=1000
    boot_strap_no_samples=100
    probab_list=[]
    for ii in range(boot_strap_no_samples):
        boot_sample=np.random.choice(arr, size=boot_strap_sample_size, replace=False)
        val,bin_v=np.histogram(boot_sample,range=(minf,maxf),bins=no_bins)
        probab=val/sum(val)
        probab_list.append(probab)
    probab_arr=np.array(probab_list)
    bin_mid=0.5*(bin_v[0:-1]+bin_v[1:])
    return bin_mid,np.mean(probab_arr,axis=0),np.std(probab_arr,axis=0)
def plot_distribution(df,typ,maxf,no_bins):
    fig,ax=plt.subplots()
    for key1 in dic_diff.keys():
        for key2 in dic_rotation.keys():
            arr=df["AS16"+key1+"H"+key2+"_v3"+"3"]
            arr=arr*1e12
            if typ=='abs_mag_only':
                print('Absolute Magnitude')
                arr=abs(arr)
                maxf=maxf
                minf=0
            else:
                minf=-maxf
                maxf=maxf
    
            bin_mid,vals,std = bootstapper(arr,minf,maxf,no_bins)
            ax.errorbar(bin_mid,vals,yerr=std,marker=dic_rotation[key2][2],ms=marker_size,color=colors_dic[key1+key2],ls=dic_rotation[key2][1],
                    ecolor="0.8",label=dic_diff[key1][0]+", "+dic_rotation[key2][0])
            
            if key1=="R":
                arr=df["AS16"+key1+"H"+key2+"_v3"+"1"]
                arr=arr*1e12
                if typ=='abs_mag_only':
                    print('Absolute Magnitude')
                    arr=abs(arr)
                    maxf=maxf
                    minf=0
                else:
                    minf=-maxf
                    maxf=maxf
            bin_mid,vals,std = bootstapper(arr,minf,maxf,no_bins)
            ax.plot(bin_mid, vals, marker=dic_rotation[key2][2], ms=marker_size, ls=':',color=colors_dic[key1+key2],
                    label=", "+dic_rotation[key2][0])
            
    ax.set_yscale("log")
    ax.set_ylim([1e-3,1.1])
    #ax.set_xlim([0,20])
    legend_elements = [Line2D([0], [0], color=colors_dic['RR'],ls='-', label='Rigid'),
                       Line2D([0], [0], color=colors_dic['D4R'],ls='-', label='Lipid'),
        Line2D([0], [0], color='k', ls='--',label='n = 3'),
                       Line2D([0], [0], color='k', ls=':',label='n = 1'),
                   Line2D([0], [0], marker=dic_rotation['R'][2], color='k',ls='none', label='With Rotation',
                          markersize=marker_size+3),
                    Line2D([0], [0], marker=dic_rotation['NR'][2], color='k',ls='none', label='Without Rotation',
                          markersize=marker_size+3),
                   ]
    if typ=='abs_mag_only':
        ax.set_xlabel('|f| (pN)',fontsize=14)
        ax.legend(handles=legend_elements,fontsize=12)
    else:
        ax.axvline(x=0,ls='-',lw=0.5,color='k')
        ax.set_xlabel('Magnitude of motor force, f (pN)',fontsize=14)
        #ax.legend(handles=legend_elements,fontsize=12)
    ax.set_ylabel('Probability',fontsize=14)
    filename=typ
    fig.savefig(filename+".svg", bbox_inches='tight')
    fig.savefig(filename+".pdf", bbox_inches='tight')
    fig.savefig(filename+".png", bbox_inches='tight',dpi=600)
    fig.savefig(filename+".eps", bbox_inches='tight',dpi=600)

plot_distribution(df,typ='abs_mag_only',maxf=11,no_bins=25)
plot_distribution(df,typ='all',maxf=11,no_bins=25)
```
