```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

```python
df=pd.read_csv('average_height_with_rotation.csv',index_col=0)
```

```python
dic_diff={0:['Rigid','gray'],1:["Lipid",'darkgreen']}
dic_rotation={1:['With rotation','-',"^"],0:["Without rotation",'--','o']}
```

```python
fig, ax = plt.subplots(figsize=(4,3))
mult=1e9
for key_diff in dic_diff:
    df_ext1=df.loc[df["D"]==key_diff]
    for key_rot in dic_rotation.keys():
        df_ext=df_ext1.loc[df_ext1["Rotation"]==key_rot]
        ax.errorbar(df_ext["no_bound"],mult*df_ext["average_height"],yerr=mult*df_ext["error"],
                    color=dic_diff[key_diff][1],
                    ls=dic_rotation[key_rot][1],
                    marker=dic_rotation[key_rot][2],
                
                label= dic_diff[key_diff][0]+", "+dic_rotation[key_rot][0])
ax.set_xlabel(r'Number of bound motors, $n$', fontsize=13)
ax.set_ylabel(r'Average height, H $(nm)$', fontsize=13)
ax.legend(fontsize=10)
steric_line=12.5+250
ax.axhline(y=steric_line,ls=':',color='k')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
filename='height'
fig.savefig(filename+".svg", bbox_inches='tight')
fig.savefig(filename+".pdf", bbox_inches='tight')
fig.savefig(filename+".eps", bbox_inches='tight',dpi=600)
fig.savefig(filename+".png", bbox_inches='tight',dpi=600)
```
